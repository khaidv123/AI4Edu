````python
Prompt_classify = '''
Cho một câu trả lời tương tác của của AI trong đoạn chat , nhiệm vụ của bạn như sau:

## Phân loại:
    "question" : ở đây nghĩa là AI chỉ có hỏi yêu cầu người khác trả lời một kiến thức hay làm gì đó nào đó.
    "hint" : đưa ra gợi ý, mặc dụ có đặt câu hỏi nhưng vẫn có thông tin hữu ích kích thích suy nghĩ người khác mà không phải hỏi suông.
    "opinion" : đưa ra ý kiến với người khác.

## Đếm số hành động:
    Đếm xem trong câu trả lời AI thực hiện bao nhiêu hành động, ví dụ nêu ý kiến tính là một hành động, nêu 1 câu hỏi là 1 hành động, nêu 2 câu hỏi là 2 hành động,... Và cộng tổng lại xem bao nhiêu hành động.

Đầu ra JSON:
---
```json
{
    "classify" : <question/hint/opinion>,
    "num" : "<số hành động>"
}
```
---
Ví dụ:
---
```json
{  
    "classify" : "opinion",
    "num" : "1"
}
```
---

Đây là câu nói của AI:
<talk>
'''
````

````python
import json
import time

from openai import OpenAI
import os

from llm import *
from prompts import *

client = OpenAI(
    base_url="https://api.studio.nebius.com/v1/",
    api_key=os.environ.get("NEBIUS_API_KEY")
)

def llm(prompt, tem = 0):
    response = client.chat.completions.create(
            model="google/gemma-3-27b-it",
            max_tokens=2048,
            temperature= tem,
            top_p=0.9,
            messages=[ {
                "role": "user",
                "content": [
                    {"type" : "text", "text" : prompt},
                ]
            }],
        )
    return response.choices[0].message.content
````

````python
Bob = []
Alice = []
Charlie = []
````

````python
for t in HISTORY:
    if t[0] != "Khải":
        name = t[0]
        text = t[1]

        a = llm(Prompt_classify.replace("<talk>", text))
        a = json.loads(a.strip('```json').strip('\n```'))

        if name == "Bob":
            Bob.append(a)
        elif name == "Alice":
            Alice.append(a)
        elif name == "Charlie":
            Charlie.append(a)
        else:
            print("no name")
            break
````

````python
num_Bob = []
classify_Bob = []
for i in Bob:
    num_Bob.append(int(i['num']))
    classify_Bob.append(i['classify'])
````

````python
import matplotlib.pyplot as plt
from collections import Counter
# Đếm số lần xuất hiện
counter = Counter(num_Bob)

# Tạo biểu đồ
plt.bar(counter.keys(), counter.values(), color=['skyblue', 'orange', 'green'])
plt.xlabel('Giá trị')
plt.ylabel('Số lần xuất hiện')
plt.title('Biểu đồ tần suất xuất hiện các số')
plt.xticks([1, 2, 3])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hiển thị biểu đồ
plt.show()
````
